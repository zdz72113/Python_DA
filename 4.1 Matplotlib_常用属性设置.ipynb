## 常用属性设置

Matplotlib的一些常用属性设置

### Agenda

- 图像和子图
- 坐标
- 标题和图例
- 线条属性
- 文本和注释

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 面板和子图

```python
#创建测试数据
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

print(x[0:10])
print(y[0:10])
```

创建图形. 一些常用参数：num为窗口的属性id，figsize设置窗口大小，dpi设置窗口分辨率等。

```python
plt.figure(num=1, figsize=(16, 9), dpi=80)
```

创建子图。参数221代表生成2行2列，当前子图为第1个。

```python
# 线性
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')

# 对数
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.05)
plt.title('symlog')

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')

plt.show()
```

### 坐标

- xlim():设置x坐标轴范围
- ylim():设置y坐标轴范围
- xlabel():设置x坐标轴名称
- ylabel():设置y坐标轴名称
- xticks():设置x轴刻度
- yticks():设置y轴刻度

```python
#创建测试数据
x = np.linspace(-5, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(num=2, figsize=(8, 5))
plt.plot(x, y1)
plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='--')

#设置坐标轴范围
plt.xlim((-5, 5))
plt.ylim((-2, 2))

#设置坐标轴名称
plt.xlabel('X')
plt.ylabel('Y')

#设置坐标轴刻度
plt.xticks(np.arange(-5, 5, 0.5))
plt.yticks(np.arange(-2, 2, 0.3))

plt.show()
```

- gca():获取当前坐标轴信息
- spines:设置边框
- set_color:设置边框颜色：默认白色
- xaxis.set_ticks_position:设置x坐标刻度数字或名称的位置
- yaxis.set_ticks_position:设置y坐标刻度数字或名称的位置
- set_position:设置边框位置

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3],['$really\ bad$', '$bad$', '$normal$', '$good$', '$really\ good$'])

#获取坐标轴
ax = plt.gca()

#设置上边和右边无边框
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#设置x坐标刻度数字或名称的位置
ax.xaxis.set_ticks_position('bottom')
#设置边框位置
ax.spines['bottom'].set_position(('data', 0))

#设置x坐标刻度数字或名称的位置
ax.yaxis.set_ticks_position('left')
#设置边框位置
ax.spines['left'].set_position(('data',0))

plt.show()
```

### 标题和图例

设置标题

```python
#创建测试数据
x = np.linspace(-5, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(num=2, figsize=(8, 5))
plt.plot(x, y1)
plt.plot(x, y2)

#设置标题
plt.title('This is the title',fontsize='large',fontweight='bold',color='blue',loc ='left')

plt.show()
```

设置图例

```python
#创建测试数据
x = np.linspace(-5, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(num=2, figsize=(8, 5))
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')

#显示图例
#plt.legend()

#图例及位置： 0最优 1右上角，2左上角 ； ncol为一列有几个标签
plt.legend(loc=2, ncol=2)

plt.show()
```

### 线条属性

设置线条的颜色（Color）、样式（Style）以及记号（Marker）等

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2.3, 3.4, 1.2, 6.6, 7.0]
plt.scatter(x, y, color='r', marker='+')
plt.plot(x, y, color='blue', linestyle='--')

plt.show()
```

### 文本和注释

使用annotate方法标注文字

```python
#创建测试数据
x = np.arange(0, 6)
y = x * x

plt.plot(x, y, marker='o')
for xy in zip(x, y):
    plt.annotate("(%s,%s)" % xy, xy=xy, xytext=(-20, 10), textcoords='offset points')
plt.show()
```

使用text方法显示文字

```python
plt.plot(x, y, marker='o')
plt.text(2,15,"$y=x^2$",fontsize=15,verticalalignment="bottom",horizontalalignment="left")
plt.show()
```

使用annotate方法种的arrowprops参数显示箭头

```python
plt.plot(x, y, marker='o')
plt.annotate("(%s,%s)" % xy, xy=(3, 9), xytext=(-20, 30), arrowprops=dict(facecolor='black', shrink=0.05), textcoords='offset points')
plt.show()
```
