## 列的基本处理

类似于Excel和数据库表，我们怎么增加，修改，删除，选择列

### Agenda

- 增加
- 删除
- 修改
- 选择

```python
import pandas as pd
```

### 增加

增加一列，用df['新列名'] = 新列值，在原数据基础上赋值即可：

```python
#读取csv文件
imdb = pd.read_csv('Data/imdb_1000.csv')

#增加新列title_len
imdb['title_len'] = imdb['title'].str.len() + 1

#显示前5行
imdb.head(5)
```

### 删除

用drop函数指定删除的列，axis = 1表示针对列的操作，inplace为True，则直接在源数据上进行修改，否则源数据会保持原样。

```python
imdb.drop('title_len', axis = 1, inplace = True)

imdb.head(5)
```

### 修改

最简单的修改：df['原列名'] =  值，就完成了对原列数值的修改。

```python
# duration从分钟修改为小时
imdb['duration'] =  round(imdb['duration'] / 60, 2)

# 显示前5行
imdb.head(5)
```

### 选取

```python
# 选取单行
imdb['title']
```

```python
# 选取多行
imdb[['title', 'star_rating']]
```
