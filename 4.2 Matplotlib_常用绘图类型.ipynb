## 常用绘图类型

Matplotlib的一些常用绘图类型

### Agenda

- 条形图
- 散点图
- 折线图
- 面积图
- 直方图
- 箱线图
- 热力图
- 雷达图
- 饼图

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 条形图

利用矩阵条的高度反映数值变量的集中趋势，以及估计变量之间的差值统计（置信区间）。

```python
#创建测试数据
x = [52, 69, 58, 12, 39, 75]
y = [56, 15, 84, 65, 45, 48]
index = np.arange(len(x))

width_val = 0.4 #若显示 n 个柱状图，则width_val的值需小于1/
plt.bar(index, height=x, width=width_val, alpha=0.5, color='red', label='x')
plt.bar(index+width_val, height=y, width=width_val, alpha=0.5, color='green', label='y')

plt.legend()
plt.show()
```

水平条形图

```python
#创建测试数据
x = [52, 69, 58, 12, 39, 75]
y = [56, 15, 84, 65, 45, 48]
index = np.arange(len(x))

height_val = 0.4 #若显示 n 个柱状图，则width_val的值需小于1/
plt.barh(index, x, height=height_val, alpha=0.5, color='red', label='x')
plt.barh(index+height_val, y, height=height_val, alpha=0.5, color='green', label='y')

plt.legend()
plt.show()
```

### 散点图

体现变量间相互影响程度和相关性

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```

```python
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5) 
plt.colorbar()
plt.show()
```

### 折线图

显示数据的变化趋势

```python
#创建测试数据
x = np.arange(0, 6)
y1 = x * x
y2 = x ** 3

plt.plot(x, y1, marker='o', label='y1')
plt.plot(x, y2, marker='*', label='y1')

plt.legend()
plt.show()
```

### 面积图

面积图，既反应单个变量的变化趋势，也反应总量趋势

```python
#创建测试数据
x = np.arange(0, 6)
y1 = x * x
y2 = x ** 3

plt.stackplot(x, y1,y2, labels=['y1','y2'])

plt.legend(loc=2)
plt.show()
```

### 直方图

反应数据的分别规律

```python
data = np.random.randn(10000)
"""
绘制直方图
data:必选参数，绘图数据
bins:直方图的长条形数目，可选项，默认为10
density:是否将得到的直方图向量归一化，可选项，默认为0，代表不归一化，显示频数。normed=1，表示归一化，显示频率。
facecolor:长条形的颜色
edgecolor:长条形边框的颜色
alpha:透明度
"""
plt.hist(data, bins=40, density=0, facecolor="blue", edgecolor="black", alpha=0.7)

plt.show()
```

### 箱线图

箱线图一般用来展现数据的分布（如上下四分位值、中位数等），同时，也可以用箱线图来反映数据的异常情况。

```python
data = (np.array([1908.3, 3158.2, 4140.6, 5510.2, 2015.3, 3235.0, 4453.8, 5798.4,
               2147.6, 3385.8, 4731.2, 5925.6, 2222.5, 3447.2, 5046.1, 12000.4]),
     np.array([9548.0, 11127.5, 11887.0, 13102.3, 10641.7, 12312.9, 12790.3,
               13915.8, 11320.0, 13300.1, 14024.3, 15461.0, 13146.6, 15219.9]),
     np.array([9873.6, 9757.7, 9684.9, 10581.7, 11429.4, 11178.6, 11089.3, 12002.6,
               12827.3, 12508.9, 12501.8, 13583.8, 14456.4, 13870.2, 13946.9])
     )
labels = ["a", "b", "c"]

plt.boxplot(x = data, # 指定绘图数据
            patch_artist=True, # 要求用自定义颜色填充盒形图，默认白色填充
            showmeans=True, # 以点的形式显示均值
            boxprops = {'color':'black','facecolor':'#9999ff'}, # 设置箱体属性，填充色和边框色
            flierprops = {'marker':'o','markerfacecolor':'red','color':'black'}, # 设置异常值属性，点的形状、填充色和边框色
            meanprops = {'marker':'D','markerfacecolor':'indianred'}, # 设置均值点的属性，点的形状、填充色
            medianprops = {'linestyle':'--','color':'orange'}) # 设置中位数线的属性，线的类型和颜色

plt.show()
```

### 热力图

热力图用颜色的深浅表示数据对应位置值的大小。

```python
x = np.random.rand(10, 10)

#使用seaborn热力图绘制函数heatmap
sns.heatmap(x, # 指定绘图数据
            cmap=plt.cm.Blues, # 指定填充色
            linewidths=.1, # 设置每个单元方块的间隔
            annot=True # 显示数值
            )

plt.show()
```

### 雷达图

雷达图刻画出某些指标的横向或纵向的对比关系

```python
data = [0.8, 0.9, 1.2, 1.0, 1.5, 0.8]

criterion = [1, 1, 1, 1, 1, 1] # 基准雷达图
angles = np.linspace(0, 2 * np.pi, 5, endpoint=False)
angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure(facecolor='#87CEEB') # 创建画板并填充颜色
ax = fig.add_subplot(111, polar=True,)  # 设置坐标为极坐标

# 绘制三个五边形
floor = 0
ceil = 2
labels = np.array(['x1', 'x2', 'x3', 'x4', 'x5'])

# 绘制五边形的循环
for i in np.arange(floor, ceil + 0.5 ,0.5):
    ax.plot(angles, [i] * (6), '-', lw= 0.5, color='black')
for i in range(5):
    ax.plot([angles[i], angles[i]], [floor, ceil], '-',lw=0.5, color='black')

# 绘制雷达图
ax.plot(angles, criterion, 'b-', lw=2, alpha=0.4)
ax.fill(angles, criterion, facecolor='b', alpha=0.3) #填充
ax.plot(angles, data, 'b-', lw=2, alpha=0.35)
ax.fill(angles, data, facecolor='b', alpha=0.25)

ax.set_thetagrids(angles * 180 / np.pi, labels)
ax.spines['polar'].set_visible(False)#不显示极坐标最外的圆形
ax.set_theta_zero_location('N')#设置极坐标的起点（即0度）在正上方向
ax.grid(False)# 不显示分隔线
ax.set_yticks([]) # 不显示坐标间隔
ax.set_title('雷达图', va='bottom', fontproperties='SimHei')
ax.set_facecolor('#87ceeb') # 填充绘图区域的颜色

plt.subplots_adjust(left=0.09, right=1, wspace=0.25, hspace=0.25, bottom=0.13, top=0.91)
plt.show()
```

```python
饼图体项了各个部分构成比例的情况
```

```python
labels = ["x", "y", "z"]    # 各部分标签
size = [55, 30, 15]    # 各部分大小
color = ["steelblue", "green", "blue"]     # 各部分颜色
explode = [0.05, 0, 0]   # 各部分突出值
"""
绘制饼图
explode：设置各部分突出
label:设置各部分标签，
labeldistance:设置标签文本距圆心位置，1.1表示1.1倍半径
autopct：设置圆里面文本
shadow：设置是否有阴影
startangle：起始角度，默认从0开始逆时针转
pctdistance：设置圆内文本距圆心距离
"""
plt.pie(size, explode=explode, colors=color, labels=labels, 
                                  labeldistance=1.1, autopct="%1.1f%%", shadow=True, startangle=90, pctdistance=0.6)

plt.legend()
# 将横、纵坐标轴标准化处理，保证饼图是一个正圆，否则为椭圆
plt.axis("equal")
plt.show()
```
